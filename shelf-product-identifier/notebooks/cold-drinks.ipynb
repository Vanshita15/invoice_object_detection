```python
# Import cc th vin
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import torch
import pandas as pd

# Bc 1: Ci t yolov8
!pip install yolov8
import yolov8

# Bc 2: To tp data.yaml
data_yaml = """
train: /kaggle/input/cold-drinks-inventory-dataset/Finalize/images/train
val: /kaggle/input/cold-drinks-inventory-dataset/Finalize/images/test
nc: {}  # S lp
names: {}  # Tn lp
"""

# c classes.txt
with open("/kaggle/input/cold-drinks-inventory-dataset/classes.txt", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Cp nht data.yaml
data_yaml = data_yaml.format(len(classes), classes)

# Lu data.yaml
with open("/kaggle/working/data.yaml", "w") as f:
    f.write(data_yaml)

print("To tp data.yaml thnh cng:")
print(data_yaml)

# Bc 3: Hun luyn M hnh yolov8
def train_yolov8():
    """
    Hun luyn m hnh yolov8s trn tp d liu.
    """
    !python -m yolov8.train \
        --img 640 \
        --batch 16 \
        --epochs 20 \
        --data /kaggle/working/data.yaml \
        --weights yolov8s.pt \
        --cache \
        --project /kaggle/working/yolov8 \
        --name inventory_training

# Bc 4: Pht hin Sn phm
def detect_products(image_path, model, class_names):
    """
    Pht hin sn phm trong hnh nh bng yolov8.
    Tr v danh sch (bbox, label, confidence).
    """
    # c hnh nh
    img = cv2.imread(image_path)
    if img is None:
        return [], img
    
    # Chy suy lun
    results = model(image_path)
    
    # Ly kt qu
    detections = []
    pred = results.pred[0]  # D on cho hnh nh
    for *box, conf, cls in pred:
        x1, y1, x2, y2 = map(int, box)
        label = class_names[int(cls)]
        detections.append(([x1, y1, x2-x1, y2-y1], label, conf.item()))
    
    return detections, img

# Bc 5: Cp nht Tn kho
def update_inventory(detections, inventory_db):
    """
    Cp nht mc tn kho da trn sn phm c pht hin.
    """
    for _, label, _ in detections:
        if label in inventory_db:
            inventory_db[label] += 1
        else:
            inventory_db[label] = 1
    return inventory_db

# Bc 6: Trc quan ha
def visualize_detections(image, detections):
    """
    Hin th hnh nh vi hp gii hn v nhn.
    """
    img = image.copy()
    for bbox, label, confidence in detections:
        x, y, w, h = bbox
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, f"{label} ({confidence:.2f})", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Sn phm c Pht hin")
    plt.show()

def plot_inventory(inventory_db):
    """
    V biu  tn kho bng Seaborn.
    """
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(inventory_db.values()), y=list(inventory_db.keys()))
    plt.title("Mc Tn kho Hin ti")
    plt.xlabel("S lng")
    plt.ylabel("Sn phm")
    plt.show()

# Bc 7: Hm Chnh
def main():
    # ng dn tp d liu
    dataset_path = "/kaggle/input/cold-drinks-inventory-dataset/Finalize"
    
    # Hun luyn m hnh
    print("Bt u hun luyn yolov8...")
    train_yolov8()
    
    # Ti m hnh  hun luyn
    model_path = "/kaggle/working/yolov8/inventory_training/weights/best.pt"
    if not os.path.exists(model_path):
        print("Khng tm thy m hnh  hun luyn. S dng yolov8s.pt mc nh.")
        model = torch.hub.load('ultralytics/yolov8', 'yolov8s', pretrained=True)
    else:
        model = torch.hub.load('ultralytics/yolov8', 'custom', path=model_path)
    
    # Khi to c s d liu tn kho
    inventory_db = {}
    
    # Th mc hnh nh kim tra
    test_images_path = os.path.join(dataset_path, "images/test")
    image_files = list(Path(test_images_path).glob("*.jpg"))
    
    # M phng pht hin theo thi gian thc
    for img_path in image_files[:5]:  # Gii hn 5 hnh nh  chy nhanh
        detections, image = detect_products(str(img_path), model, classes)
        if not detections:
            print(f"Khng pht hin sn phm trong {img_path}")
            continue
        
        # Cp nht tn kho
        inventory_db = update_inventory(detections, inventory_db)
        
        # Trc quan ha
        visualize_detections(image, detections)
        plot_inventory(inventory_db)
    
    # Lu kt qu tn kho
    pd.DataFrame.from_dict(inventory_db, orient='index', columns=['Quantity']).to_csv('/kaggle/working/inventory.csv')
    print(" lu kt qu tn kho vo /kaggle/working/inventory.csv")

# Chy chng trnh
if __name__ == "__main__":
    main()
```
